### **Base de ventas**

> #### <strong><font color="7b71ff" size =6>Prompt:</font></strong>

Ahora vamos explorar la base de ventas en formato de un DataFrame, llamado "ventas". Contiene datos de las ventas del comercio electrónico, que son: ID de la compra, fecha de la compra, hora de la compra, categoría del producto, precio unitario del producto, cantidad vendida, envío y el método de pago.

Queremos hacer una breve exploración de los datos para comprender un poco sobre ellos. Recordando que los datos de esta base también están limpios y tratados. Construye el código para ejecutar esta etapa utilizando Python con el objetivo de conocer mejor el dataframe "clientes" y algunos de sus atributos.

```python
# Dimensiones del DataFrame
ventas.shape

# Estadísticas descriptivas incluyendo variables categóricas
ventas.describe(include="object")

# Verificación de valores nulos por columna
ventas.isna().sum()

# Distribución por categoría de producto
ventas["categoria_producto"].value_counts()

# Distribución por método de pago
ventas["metodo_pago"].value_counts()

# Estadísticas del precio unitario
ventas["precio_unitario"].describe()

# Estadísticas de la cantidad vendida
ventas["cantidad"].describe()

# Cálculo del valor total por venta
ventas.assign(valor_total=ventas["precio_unitario"] * ventas["cantidad"]).head()

# Valor promedio de venta por categoría
ventas.assign(valor_total=ventas["precio_unitario"] * ventas["cantidad"]) \
      .groupby("categoria_producto")["valor_total"].mean()

# Número de compras únicas
ventas["id_compra"].nunique()

# Cantidad total de productos vendidos
ventas["cantidad"].sum()

# Ventas por método de pago (en valor monetario)
ventas.assign(valor_total=ventas["precio_unitario"] * ventas["cantidad"]) \
      .groupby("metodo_pago")["valor_total"].sum()
```

```python
# Vista general de los primeros registros
ventas.head()
```

```python
# Información general: columnas, tipos de datos y valores no nulos
ventas.info()
```

```python
# Conversión de la columna fecha a datetime
ventas["fecha"] = pd.to_datetime(ventas["fecha"], format='%Y-%m-%d')
```

```python
# Estadísticas descriptivas de variables numéricas
ventas.describe()
```

Ahora vamos a unir las dos bases de nuestro proyecto en una sola utilizando como llave la columna "ID_compra". Además, organizaremos las columnas en la siguiente secuencia:

* ID_compra

* cantidad

* metodo_pago
* ID_cliente

Construye un código para ejecutar esta etapa usando Python y la biblioteca Pandas.


```python
# Unión de las bases de datos usando ID_compra como llave
df = ventas.merge(
    clientes,
    on="ID_compra",
    how="inner"
)
```

```python
# Reordenamiento de las columnas según el orden solicitado
df = df[
    [
        "ID_compra",
        "fecha",
        "horario",
        "categoria",
        "precio_unitario",
        "cantidad",
        "envio",
        "metodo_pago",
        "ID_cliente",
        "edad",
        "sexo_biologico",
        "ciudad",
        "estado",
        "region",
        "cashback",
        "nota"
    ]
]
```

```python
# Vista rápida del resultado
df.head()
```

## 1️⃣ Análisis de facturación y desempeño en el tiempo

**Objetivo:** entender estacionalidad, tendencias y picos de venta.

### 📈 Serie temporal de facturación

* **Tipo:** Línea
* **Qué muestra:** evolución de la facturación a lo largo del año
* **Variables:**

  * Eje X → `fecha` (agrupada por mes)
  * Eje Y → facturación (`precio_unitario * cantidad + envio`)
* **Insight esperado:** meses fuertes/débiles, tendencia general

### 📊 Facturación por mes

* **Tipo:** Barras
* **Variables:** mes vs facturación total
* **Insight:** comparación clara entre períodos

## 2️⃣ Análisis por categoría de producto

**Objetivo:** identificar qué productos generan más ingresos y volumen.

### 📊 Facturación por categoría

* **Tipo:** Barras horizontales
* **Variables:** `categoria` vs facturación
* **Insight:** categorías estratégicas para el negocio

### 📦 Volumen vendido por categoría

* **Tipo:** Barras
* **Variables:** `categoria` vs `cantidad`
* **Insight:** diferencia entre productos de alto volumen y alto valor
